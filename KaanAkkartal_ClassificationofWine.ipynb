## 2.3. Data Preparation

```python
df.dtypes
```

```python
df["class"] = pd.Categorical(df["class"])
df.dtypes
```

```python
df.head()
```

```python
X = df.drop(columns = ["class", "flav", "phenols"])
X.head()
```

```python
Y = df["class"]
Y.head()
```

```python
Y = pd.get_dummies(Y, drop_first=True)
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X , Y, test_size = 0.30)
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 3. Modelling

## 3.1. Application, Prediction and Evaluation 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

cart_model = DecisionTreeClassifier().fit(X_train_scaled, Y_train)
rf_model = RandomForestClassifier().fit(X_train_scaled, Y_train)
mlpc = MLPClassifier().fit(X_train_scaled, Y_train)
```

```python
Y_pred_cart = cart_model.predict(X_test_scaled)
Y_pred_rf = rf_model.predict(X_test_scaled)
Y_pred_mlpc = mlpc.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cart_model_score = accuracy_score(Y_test, Y_pred_cart) 
rf_model_score = accuracy_score(Y_test, Y_pred_rf) 
mlpc_score = accuracy_score(Y_test, Y_pred_mlpc) 

print("Decision tree accuracy score is: " , cart_model_score)
print("Random forest accuracy score is: " , rf_model_score)
print("Neural network accuracy score is: " , mlpc_score)
```

```python
print("                           Decision Tree " + "\n" + classification_report(Y_test, Y_pred_cart))
print("                           Random Forest " + "\n" + classification_report(Y_test, Y_pred_rf))
print("                           Neural Network " + "\n" + classification_report(Y_test, Y_pred_mlpc))
```

## 3.2. Model Tuning - Random Forest

```python
from sklearn.model_selection import GridSearchCV
```

```python
rf_model
```

```python
rf_params = {"max_depth": [2,5,8,10],
            "max_features": [2,5,8],
            "n_estimators": [10,500,1000],
            "min_samples_split": [2,5,10]}
```

```python
rf_model = RandomForestClassifier()

rf_cv_model = GridSearchCV(rf_model, 
                           rf_params, 
                           cv = 5, 
                           n_jobs = -1, 
                           verbose = 2) 
```

```python
rf_cv_model.fit(X_train_scaled, Y_train)
```

```python
print("En iyi parametreler: " + str(rf_cv_model.best_params_))
```

```python
rf_tuned = RandomForestClassifier(max_depth = 5, 
                                  max_features = 2, 
                                  min_samples_split = 2,
                                  n_estimators = 500)

rf_tuned.fit(X_train_scaled, Y_train)
```

```python
Y_pred_tuned = rf_tuned.predict(X_test_scaled)
accuracy_score(Y_test, Y_pred_tuned)
```

```python
print("                           Random Forest Tuned " + "\n" + classification_report(Y_test, Y_pred_tuned))
```
